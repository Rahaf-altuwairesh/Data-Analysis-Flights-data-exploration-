```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
flights_df=pd.read_csv('flights_clean.csv')
df_cancelled=pd.read_csv('flights_cancelled_clean.csv')
```

## What is the percentage of canceled flights in the first quarter of 2008?
The flight cancellation rate in the first quarter of 2008 was 3%.

```python
# Set the size
axis = plt.subplots(figsize=(6,7))
Name = flights_df['Cancelled'].unique()
size = flights_df['Cancelled'].value_counts()
#cancelled pie plot
cancelled_plot=plt.pie(size,labels=Name,autopct='%5.0f%%')
# set the title
plt.title('Flight cancellation rate for the first quarter of 2008',fontsize=14)
plt.show()
```

## Which airline has the highest flights in the first quarter?
The airline with the highest frequency is WN (Southwest Airlines), and the airline with the lowest frequency AQ (Aloha Airlines).

```python
# Set the size
plt.figure(figsize=(6, 6))
#count plot for Unique Carrier
axis = sb.countplot(x=flights_df['UniqueCarrier'], data = flights_df,
              order=flights_df['UniqueCarrier'].value_counts().index)
axis.set_xticklabels(axis.get_xticklabels(), rotation=90, ha="right")
plt.tight_layout()
# set the title
plt.title('Unique carrier code frequency',fontsize=14)
plt.xlabel('Unique Carrier Code',fontsize=13)
plt.show()

#Printing a list of airlines' names
print(['WN: Southwest Airlines', 'AA: American Airlines', 'OO: Skywest Airlines','MQ: American Eagle Airlines',
       'UA: United Airlines','US: US Airways','DL: Delta Airlines','XE: ExpressJet','NW: Northwest Airlines',
       'CO: Continental Airlines','EV: Atlantic Southeast Airlines','9E: Pinnacle Airlines',
       'FL: AirTran Airways','YV: Mesa Airlines','OH: Comair','B6: JetBlue Airways',
       'AS: Alaska Airlines','F9: Frontier Airlines','HA: Hawaiian Airlines','AQ: Aloha Airlines'])
```

## What is the shape of the distribution of delay departure? Are there any interesting values? What is the distribution of the most values?
Delayed departure has outliers however most of the values were around and at zero. and By seeing the plot and skewness(6.30) and kurtosis(99.80) indicators, that delays lie mostly on the left side - most delays are short - of the graph, with a long tail to the right - longer delays cause higher right skewness.We also note interesting negative values, which indicate that the trips started ahead of schedule.

```python
# Set the size
fig = plt.figure(figsize = [20, 10])
# set the title
fig.suptitle('Departure delays distribution', fontsize=16)

plt.subplot(1, 2, 1)
# Plot the Departure delays
DepDelay_plot = sb.distplot(flights_df['DepDelay'])
# Set labels names
DepDelay_plot.set(xlabel='Departure Delays in Minutes', ylabel='Number of delays');


# plot another distribution plot including the zoom
plt.subplot(1, 2, 2)
bin_edges = np.arange(flights_df['DepDelay'].min(), flights_df['DepDelay'].max()+2, 10)
# plot the distribution plot with the bins
Dep_zoom_plot = sb.distplot(flights_df['DepDelay'], bins = bin_edges, kde = False, hist_kws = {'alpha' : 1})

# set the x values limitation for the for a magnified distribution plot. I start at the min() and use 250 as the right limit
plt.xlim((flights_df['DepDelay'].min(), 250))
# Set labels names
Dep_zoom_plot.set(xlabel='Departure Delays in Minutes', ylabel='Number of delays');
```

## What are the most common reasons for canceling flights?
The biggest reason for flight cancellations was the carrier and the weather, followed by the National Air System.

```python
# Set the size
f,ax=plt.subplots(1,2,figsize=(20,8))
#pie plot for Cancellation Code
flights_df['CancellationCode'].value_counts().plot.pie(explode=[0.05,0.05,0.05,0.05],autopct='%1.1f%%',ax=ax[0],shadow=True)
ax[0].set_ylabel('')
#count plot for Cancellation Code
sb.countplot('CancellationCode', order = flights_df['CancellationCode'].value_counts().index, data=flights_df, ax=ax[1])
plt.xlabel('Reason for cancellation',fontsize=13)
plt.show()
```

## Which airline has the highest number of flight cancellations?
- The airline AA(American Airlines)occupies the highest number of flight cancellations, and the airline AQ(Aloha Airlines)occupies the lowest number of flight cancellations,and this is logical because the number of flights in it is few.
- The interesting thing is that the airline WN(Southwest Airlines) is located in the average cancellation values, and it occupies the most number of flights.

```python
#Visualize canceled flights for each carrier
# Set the size
plt.figure(figsize=(6, 6))

axis = sb.countplot(x=df_cancelled['UniqueCarrier'],data = df_cancelled,
              order=df_cancelled['UniqueCarrier'].value_counts().index)
axis.set_xticklabels(axis.get_xticklabels(), rotation=90, ha="right")
plt.tight_layout()
# set the title
plt.title('The number of canceled flights according to the unique carrier code',fontsize=14)
plt.xlabel('Unique Carrier Code',fontsize=13)
plt.show()
#Printing a list of airlines' names
print(['WN: Southwest Airlines', 'AA: American Airlines', 'OO: Skywest Airlines','MQ: American Eagle Airlines',
       'UA: United Airlines','US: US Airways','DL: Delta Airlines','XE: ExpressJet','NW: Northwest Airlines',
       'CO: Continental Airlines','EV: Atlantic Southeast Airlines','9E: Pinnacle Airlines',
       'FL: AirTran Airways','YV: Mesa Airlines','OH: Comair','B6: JetBlue Airways',
       'AS: Alaska Airlines','F9: Frontier Airlines','HA: Hawaiian Airlines','AQ: Aloha Airlines'])
```

## Which airline has the highest average delay due to the carrier?
- The airlines with more than 30 minutes average delay are HA(Hawaiian Airlines),YV(Mesa Airlines),EV(Atlantic Southeast Airlines).
- The airlines with less than 11 minutes average delay are FL(AirTran Airways),B6(JetBlue Airways),WN(Southwest Airlines).
- The interesting thing is that the airline WN(Southwest Airlines) have a less than 11 minutes average delay, and it occupies the most number of flights.

```python
#Visualize Carrier Delay for each carrier
# Set the size
fig,ax=plt.subplots(1,2,figsize=(20,8))

order_Carrier=flights_df['UniqueCarrier'].value_counts().index
sb.barplot('UniqueCarrier','CarrierDelay', data=flights_df,ax=ax[0], order=order_Carrier )
# Set labels names
ax[0].set(title='Average Delay by Carrier', xlabel='Unique Carrier Code', ylabel='Carrier Delay in Minutes')

sb.boxplot('UniqueCarrier','CarrierDelay', data=flights_df,ax=ax[1], order=order_Carrier)
# Set labels names
ax[1].set(title='Delay Distribution by Carrier', xlabel='Unique Carrier Code', ylabel='Carrier Delay in Minutes')
plt.close(2)
plt.show()
#Printing a list of airlines' names
print(['WN: Southwest Airlines', 'AA: American Airlines', 'OO: Skywest Airlines','MQ: American Eagle Airlines',
       'UA: United Airlines','US: US Airways','DL: Delta Airlines','XE: ExpressJet','NW: Northwest Airlines',
       'CO: Continental Airlines','EV: Atlantic Southeast Airlines','9E: Pinnacle Airlines',
       'FL: AirTran Airways','YV: Mesa Airlines','OH: Comair','B6: JetBlue Airways',
       'AS: Alaska Airlines','F9: Frontier Airlines','HA: Hawaiian Airlines','AQ: Aloha Airlines'])
```

## What is the day with the least average delay for departure and arrival?
Friday has the most delays and Wednesday has the least number of delays-best day to fly-.

```python
#Visualize Average delay by weekday
flights_df.groupby('DayOfWeek').agg({'DepDelay':'mean', 'ArrDelay':'mean'}).plot.bar()
# set the title
plt.title('Average delay by weekday',fontsize=14)
# Set labels names
plt.xlabel('Day Of Week',fontsize=12)
plt.ylabel('Number of delays in Minutes',fontsize=11)
plt.show()
```

## What is the month with the least average delay in departure and arrival?
February has the most delays In the first quarter of the year, and April has the least number of delays-best month In the first quarter to fly-.

```python
#Visualize Average delay by month
flights_df.groupby('Month').agg({'DepDelay':'mean', 'ArrDelay':'mean'}).plot.bar()
# set the title
plt.title('Average delay by month',fontsize=14)
# Set labels names
plt.xlabel('Month',fontsize=12)
plt.ylabel('Number of delays in Minutes',fontsize=11)
plt.show()
```

## What does the average departure delay look like according to the days of the week and month from the first quarter of 2008?
- We notice that Friday is the most day that people having flights on in the first quarter of the year.
- Note in March, weekends have a high average delay.
- February has a high average delay, except for Thursday and Saturday.
- April has the lowest average delay in the first quarter on all days of the week except Friday.

```python
#Visualize Average departure delay by Month and day of week
# Set the size
fig = plt.figure(figsize=[23,8])
# set the title
fig.suptitle('Average departure delay by Month and day of week', fontsize=18)

ax = sb.barplot(data = flights_df, x = 'Month', y = 'DepDelay', hue = 'DayOfWeek')
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5), title = 'Day of Week')
# Set labels names
ax.set_xlabel("Departure Delays in Minutes - per month",fontsize=17)
ax.set_ylabel("Number of delays in Minutes",fontsize=17);
```

## With the heat map and relationship detection, are interesting relationships observed in the data set of interest?
There is a significant correlation between arrival delay and departure delay.high correlation between the scheduled arrival time and scheduled departure time , Which is logically expected. There is also an average correlation between carrier delay, arrival delay and departure delay. however,It looks like there aren't so many correlations between the quantitative features. But we observed so much detailed information when utilizing it with other features.

```python
# Set the size
plt.figure(figsize=[8,5])
#correlation plot
corr_flights_df = flights_df.corr().drop('Cancelled', axis=0)
corr_flights_df = corr_flights_df.drop('Cancelled', axis=1)
sb.heatmap(corr_flights_df, annot = True, fmt = '.2f', cmap = 'vlag_r', center = 0);
```
